```python
t = np.linspace(0,10,51)
t
```

```python
f = np.cos(t)
f
```

```python
plt.plot(t,f, color = 'green')
plt.title('График f(t)',  size = '20')
plt.xlabel('Значения t', size = '14')
plt.ylabel('Значения f', size = '14')
plt.grid()
plt.axis([0.5, 9.5, -2.5, 2.5])
plt.show()
```
